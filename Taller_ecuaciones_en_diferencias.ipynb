1. Una bomba de vacío elimina un tercio del aire restante en un cilindro con cada acción. Formule una ecuación que represente esta situación, ¿después de cuantas acciones hay solamente 1/1000000 de aire inicial?
- $U_0 = 1$
- $U_n=U_{n-1}*2/3$
- $U_{n-1} = U_{n-2}*2/3$
- $U_{n}=U_{n-2}*2/3*2/3$
- $U_{n}=U_{n-3}*2/3*2/3*2/3$
- $U_n = U_0 *(2/3)^n$
- $U_n = 1 *(2/3)^n$

  Para encontrar el valor de n cuando $U_n=1/1000000$ se puede despejar de la ecuación:
- $1/1000000=(2/3)^n$
- $log_{1/1000000}(1/1000000)=n*log_{1/1000000}(2/3)$
- $n=\frac{1}{log_{1/1000000}(2/3)}$
- $n=34$

14. Investigue el límite de $\frac{u_n}{u_{n+1}}$ si $u_n=u_{n-1}+2 u_{n-2}$

- $U_n=U_{n-1}+2U_{n-2}$
- $U_n-U_{n-1}-2U_{n-2}=0$
- $K=n-2$    
- $k+1=n-1$    
- $k+2=n$
- $U_{k+2}-U_{k+1}-2U_l=0$

Solución homogenea:
- $λ^{k+2}-λ^{k+1}-2λ^k=0$
- $λ^k(λ^2-λ-2)=0$
- $(λ-2)(λ+1)=0$
- $λ=2$
- $λ=-1$
- $U(n)=c_1(-1)^n+C_2(2)^n$
- $\frac{C(n))}{U(n+1}=\frac{C_1(-1)^n+C_2(2)^n}{c_1(-1)^{n+1}+C_2(2)^{n+1}}$
- $=\frac{C_1(-1)^n+C_2(2)^n}{-C_1(-1)^{n}+2C_2(2)^{n}}*\frac{(-1)^n}{(-1)^n}$
- $=\frac{C_1+C_2(-2)^n}{-C_1+2C_2(-2)^{n}}$

$\lim_{n\to∞} \frac{u(n)}{u(n+1)}=\lim_{n\to∞} \frac{C_1+C_2(-2)^n}{-C_1+2C_2(-2)^{n}}$

- n puede ser par o impar, solo soluciones enteras positivas:
- n par $\to$ $n=2k$ 
- n impar $\to$ $n=2k+1$ 
- si $n\to ∞$ entonces $k\to\infty$

Caso 1 $\to n$ par:
- $\lim_{n\to∞} \frac{C_1+C_2(-2)^{2k}}{-C_1+2C_2(-2)^{2k}}=\lim_{n\to∞} \frac{C_1+C_2(4)^k\to ∞}{-C_1+2C_2(4)^{k}\to ∞}$

Por L'Hopital:
- $=\lim_{n\to∞} \frac{Ln(4)C_2(4)^k}{2Ln(4)C_2(4)^k}=\lim_{n\to∞} 1/2 =1/2$

Caso 2 $\to n$ impar:
- $\lim_{n\to∞} \frac{C_1+C_2(-2)^{2k+1}}{-C_1+2C_2(-2)^{2k+1}}=\lim_{n\to∞} \frac{C_1-2C_2(4)^k\to ∞}{-C_1-4C_2(4)^{k}\to ∞}$

Por L'Hopital:
- $=\lim_{n\to∞} \frac{-2Ln(4)C_2(4)^k}{-4Ln(4)C_2(4)^k}=\lim_{n\to∞} 1/2 =1/2$

En ambos casos se cumple, por tanto el $\lim_{n\to∞} \frac{u(n)}{u(n+1)}=1/2$
